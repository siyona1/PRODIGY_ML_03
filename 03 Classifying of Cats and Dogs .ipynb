```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pickle
import random
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
pick_in = open('data.pickle','rb')
data = pickle.load(pick_in)
pick_in.close()
```

```python
random.shuffle(data)
features = []
labels = []
```

```python
for feature,label in data:
    features.append(feature)
    labels.append(label)
```

```python
x_train, x_test, y_train, y_test = train_test_split(features,labels,test_size=0.01)
```

```python
pick = open('model.sav','rb')
model = pickle.load(pick)
pick.close()
```

```python
prediction = model.predict(x_test)
accuracy = model.score(x_test, y_test)
categories = ['Cat', 'Dog']
i = 2 

print("Accuracy:", accuracy)
print("Prediction:", categories[prediction[i]])

mypet = x_test[i].reshape(50, 50)

resized_img = cv2.resize(mypet, (200, 200), interpolation=cv2.INTER_LINEAR)

plt.imshow(resized_img, cmap='gray')
plt.axis('off') 
plt.show()
```
